```python
import pandas as pd
```

```python
df = pd.read_csv('../data/A_FullClean_CountData.csv')
df.shape
```

```python
df_transposed = df.T
df_transposed.head()
```

```python
implants = ['TCPS', 'OM', 'BMP2', 'PT', 'mSLA', 'Titan']
for imp in implants:
    df_imp = df_transposed[df_transposed.index.str.contains(imp)]
    df_imp.columns = df_transposed.iloc[0]
    df_imp.to_csv(f'../data/countdata_{imp}.tsv', sep='\t', index=False)
```
