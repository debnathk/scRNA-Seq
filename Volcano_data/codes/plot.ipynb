# BP Plots

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

def plot_images_from_folder(image_folder):
    # Get the list of image files in the folder with "BP" in their name
    image_files = [file for file in os.listdir(image_folder) if file.lower().endswith(('.png')) and "BP" in file]
    print(f'No of files: {len(image_files)}')

    # Calculate the number of rows and columns for subplots
    num_images = len(image_files)
    num_rows = (num_images + 2) // 2  # Three images per row
    num_cols = min(num_images, 2)

    # Set font size based on the number of images
    title_font_size = max(12 - num_rows, 8)

    # Plot the images
    plt.figure(figsize=(15, num_rows * 5))
    for i, image_file in enumerate(image_files):
        image_path = os.path.join(image_folder, image_file)
        image = Image.open(image_path)

        plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(image)
        plt.title(image_file, fontsize=title_font_size)
        plt.axis("off")

    # Reduce both horizontal and vertical space between subplots
    plt.subplots_adjust(hspace=0, wspace=0)

    plt.show()

# Specify the folder containing images
image_folder = "../results/"

# Plot images with "BP" in their file name, three images per row
plot_images_from_folder(image_folder)
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

def plot_images_from_folder(image_folder):
    # Get the list of image files in the folder
    image_files = [file for file in os.listdir(image_folder) if file.lower().startswith(('merged'))]

    # Calculate the number of rows and columns for subplots
    num_images = min(len(image_files), 9)
    num_rows = (num_images + 2) // 3
    num_cols = min(num_images, 3)

    # Set font size based on the number of images
    title_font_size = max(12 - num_rows, 8)

    # Plot the images
    plt.figure(figsize=(15, num_rows * 5))
    for i, image_file in enumerate(image_files[:9]):
        image_path = os.path.join(image_folder, image_file)
        image = Image.open(image_path)

        plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(image)
        plt.title(image_file, fontsize=title_font_size)
        # Increase the font size of x-labels and y-labels
        plt.xticks(fontsize=title_font_size)
        plt.yticks(fontsize=title_font_size)
        plt.axis("off")

    # Reduce both horizontal and vertical space between subplots
    plt.subplots_adjust(hspace=0, wspace=0)

    plt.show()

# Specify the folder containing images
image_folder = "../results/"

# Plot the images in a 3 by 3 grid
plot_images_from_folder(image_folder)
```
