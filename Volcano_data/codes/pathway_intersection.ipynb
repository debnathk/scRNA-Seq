```python
import os
import pandas as pd
```

```python
# Merge the signature files - upsig, downsig, and sig
folder_path = "../Volcano_sig/"

def preprocess(group):
    csv_files = [file for file in os.listdir(folder_path) if file.endswith(f"_{group}sig.csv")]

    dfs = []
    for file in csv_files:
        df = pd.read_csv(os.path.join(folder_path, file), header=None)
        if df.shape[1] == 9:
            df = df.iloc[:, 1:]
        dfs.append(df)

    merged_df = pd.concat(dfs, ignore_index=True)
    merged_df = merged_df.rename(columns=merged_df.iloc[0])
    merged_df = merged_df[1:]

    merged_df.dropna(how='all', inplace=True)
    merged_df = merged_df[merged_df['Gene names'] != 'unkown'].reset_index(drop=True)

    # Select rows with padj < 0.1
    merged_df['padj'] = pd.to_numeric(merged_df['padj'], errors='coerce').dropna()
    merged_df = merged_df[merged_df['padj'] < 0.1].reset_index(drop=True)
    # return merged_df.head()

    # For common gene names, keep only that gene which have lower padj
    sorted_df = merged_df.sort_values(by='padj', ascending=True)
    final_df = sorted_df.drop_duplicates(subset='Gene names', keep='first').reset_index(drop=True)
    final_df.to_csv(folder_path + f'merged_{group}sig.csv', index=False)
    return final_df.head()
```
